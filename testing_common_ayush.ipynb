```python
with open('get_dict_t.txt') as f:
    lines = f.readlines()

f.close()
```

```python
stock_index = []
for i in lines[2:8]:
    temp_i = i.replace("\n","")
    stock_index.append(temp_i[3:])
```

```python
stock_index_line_count = []
for line,count_line in zip(lines,range(len(lines))):
    temp_line = line.replace("\n","")
    if temp_line in stock_index:
        stock_index_line_count.append(count_line)
```

```python
company_dict = {}
for line_count,indices in zip(stock_index_line_count,stock_index):
    
    index, stock_yahoo=indices.split(" ")
    print(index,stock_yahoo)

    for company in range(1,11):
        line_values = lines[line_count + company].replace("\n","")
        line_values = line_values.replace("\t","")
        
        company_stock = line_values[:line_values.find(" ")]
        company_name =  line_values[line_values.find(" ")+1:]
        
        if stock_yahoo != "No":
            company_dict[str(company_stock)+"."+stock_yahoo] = company_name
        else:
            company_dict[str(company_stock)] = company_name
```
